```python
def median(values):
    mid = (len(values)//2) - 1
    if(len(values)%2==0):
        return (values[mid] + values[mid + 1]) / 2.0
    else:
        return values[mid]
```

```python
def median_filter(values):
    values.sort()
    return median(values)
```

```python
def median_stack_color(images):
    rows = images[0].shape[0]
    cols = images[0].shape[1]
    num_channels = images[0].shape[2]
    output = np.zeros(images[0].shape)
    values = list()
    filteredValue = np.zeros(3)
#     p
    for i in range(rows):
        for j in range(cols):
#             p = (j,i)
            for channel in range(num_channels):
                values.clear()
                for k in range(len(images)):
                    values.append(images[k][i][j][channel])
                filteredValue[channel] = median_filter(values)
            output[i][j] = filteredValue
    return output
```

```python
path = "C:/Users/Kabir Juenja/Desktop/CV Research/Museum/"
images = list()
for filename in listdir(path):
    img = cv2.imread(path + filename)
    print(img.shape)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    cv2.resize(img, (1024,768), cv2.INTER_LINEAR)
    images.append(img)
filteredCvtImage = median_stack_color(images)
# print(filteredCvtImage.shape)
filteredCvtImage = filteredCvtImage.astype('uint8')
filteredCvtImage = cv2.cvtColor(filteredCvtImage,cv2.COLOR_HSV2BGR)
plt.imshow(filteredCvtImage)
cv2.imshow('Movement free Image',filteredCvtImage )
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
# print(len(images))
```
